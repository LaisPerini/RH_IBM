Importando Libs

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mit
import seaborn as sns
import warnings
```

Importando base de dados

```python
base = pd.read_csv('./Base/RH_IBM.csv',sep=',')
#base.shape                      #shape mostra quantas linhas,quantas colunas
#base.head                        #head = ultimas 5 linhas e as 5 primeiras linhas
#base.info()                           #estrutura sql mostra o tipo de dado
base.columns()                    #
```

Renomeando Colunas 

```python
base.rename(columns={
   'Age': 'Idade', 
   'Attrition': 'Atrito', 
   'BusinessTravel': 'Viagem_de_negócios', 
   'DailyRate': 'Diária', 
   'Department': 'Departamento',
   'DistanceFromHome': 'Distância_de_casa', 
   'Education': 'Educação', 
   'EducationField': 'Campo_de_educação', 
   'EmployeeCount': 'Contagem_de_empregados',
   'EmployeeNumber': 'Números_de_empregados', 
   'EnvironmentSatisfaction': 'Satisfação_ambiental', 
   'Gender': 'Gênero', 
   'HourlyRate': 'Taxa_horária',
   'JobInvolvement': 'Envolvimento_do_trabalho', 
   'JobLevel': 'Nível_de_cargo', 
   'JobRole': 'Cargo', 
   'JobSatisfaction': 'Satisfação_no_trabalho',
   'MaritalStatus': 'Estado_civil', 
   'MonthlyIncome': 'Renda_mensal', 
   'MonthlyRate': 'Taxa_mensal', 
   'NumCompaniesWorked': 'Número_de_empresas',
   'Over18': 'Mais_de_18', 
   'OverTime': 'Overtime', 
   'PercentSalaryHike': '%_salário', 
   'PerformanceRating': 'Desempenho',
   'RelationshipSatisfaction': 'Relacionamento', 
   'StandardHours': 'Horas_padrão', 
   'StockOptionLevel': 'Estoque_nível',
   'TotalWorkingYears': 'Tot_ano_trabalhado', 
   'TrainingTimesLastYear': 'Temp_Treinamento_último_ano', 
   'WorkLifeBalance': 'Equi_vida_pessoal',
   'YearsAtCompany': 'Anos_na_empresa', 
   'YearsInCurrentRole': 'Anos_func_atual', 
   'YearsSinceLastPromotion': 'Anos_última_promo',
   'YearsWithCurrManager': 'Anos_com_gerente_atual'   
    }, inplace=True)                              # para alterar e continuar na mesma base
```

Análise Univariada

```python
plt.hist(base.Idade)
plt.title('Histograma de Idade')
plt.show()
```

Relação Homem - Mulher

```python
base.Gênero.value_counts(1).plot.pie()
plt.show()
```

Funcionários por campo educacional

```python
base.Campo_de_educação.value_counts(1).plot.pie()
plt.ylabel('')
plt.show()
```

Distribuição Salarial por Gênero

```python
sns.boxplot(x='Gênero', y='Renda_mensal',data=base)
plt.show()
```

Análise de Correlação

```python
base_corr = base.select_dtypes(exclude='object')
#base.corr().style.background_gradient(cmap='Blues')

fig,ax = plt.subplots(figsize=(24,12))
sns.heatmap(base_corr.corr(),cmap= sns.cubehelix_palette(as_cmap = True),linewidth=0.5,annot=True)
```
